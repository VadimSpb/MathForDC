```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
x = np.linspace(-5, 5, 1000)
rcParams['figure.figsize'] = 5, 5

for i in range(1, 4):
    y = i*np.cos(x-i) + i
#y = k*np.cos(x-a) + b
    print(i)
    plt.grid(True)
    plt.plot(x, y)
```

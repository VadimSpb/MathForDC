```python
# графически проверяем на наличие решений:
from scipy.optimize import fsolve

def equations(i):
    x, y = i
    return ((y -x**2 + 1, np.exp(x) + x-y*x-1))


x = np.linspace(-5, 5, 1000)
plt.xlim(-2,4)
plt.ylim(-10,10)
plt.grid(True)
plt.plot(x, (np.exp(x) + x*(1-2*x-1)-1))
plt.plot(x, 2*x - 1)

x1,y1 = fsolve(equations, (-2, -1))
print(x1, y1)

x2,y2 = fsolve(equations, (2,4))
print(x2, y2)

x3,y3 = fsolve(equations, (0.1, 2))
print(x3, y3)
```

### 3,4 Дополните код Монте-Карло последовательности независимых испытаний расчетом соответствующих вероятностей (через биномиальное распределение) и сравните результаты.

```python
import itertools

k, n = 0, 10000
a = np.random.randint(0, 2, n)
b = np.random.randint(0, 2, n)
c = np.random.randint(0, 2, n)
d = np.random.randint(0, 2, n)

x = a + b + c + d

for i in range (0, n):
    if x[i] == 2:
        k += 1
#print(a, b, c, d)
#print(x)
print(k, n, k/n)
ver1 = k/n
```

```python
import math 

def Bernulli(n, k):
    p = 0.5
    q = 1 - p
    C = math.factorial(n) / (math.factorial(k) * math.factorial(n - k))
    Pn = C * (p ** k) * (q ** (n - k))
    return 'Результат по формуле Бернулли:', Pn

print('Результат по методу Монте-Карло:', ver1)
print(Bernulli(4,2))
print(Bernulli(10,3))
print(Bernulli(40,20))
```
